```python
# PLOTS WITH FIXED METRICS (one by one)

# Convert dictionary to list of tuples
data_tuples = [(k.split('--')[0], k.split('--')[1], v) for k, v in data.items()]

# Create dataframe
df = pd.DataFrame(data_tuples, columns=["Model", "Metric", "Score"])

metrics = df["Metric"].unique()  # get all unique metrics

# Create a separate bar plot for each metric
for metric in metrics:
    plt.figure(figsize=(8, 6))
    subset = df[df["Metric"] == metric]  # filter data for the current metric
    sns.barplot(data=subset, x="Model", y="Score")
    # plt.title('Metric: {}'.format(metric))
    plt.title(metric)
    plt.ylabel('Score')
    plt.xlabel('Model')
    plt.xticks(rotation=45)
    plt.show()
```
